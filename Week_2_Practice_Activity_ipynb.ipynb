```python
import pandas as pd

#getting data from csv file
df = pd.read_csv('/content/drive/MyDrive/2ndYR_Term2/MO-IT128 - Data Mining Principles/Data/nutrition_data.csv')
```

```python
# 2. Using Python, convert the categorical features into numerical format.

# One-hot encoding the meal_time column
df_transformed = pd.get_dummies(df, columns=['meal_time'], drop_first=True)

# Display the first few rows to check the transformation
df_transformed.head()
```

```python
# 3. Engineer new features you believe would be beneficial, then apply normalization or standardization techniques to the features once completed.

# Feature Engineering

# Calorie-to-protein ratio
df_transformed['calorie_protein_ratio'] = df_transformed['calories'] / (df_transformed['protein'] + 1e-6)  # Avoid division by zero

# Calorie density feature (calories per total grams of macronutrients)
df_transformed['calorie_density'] = df_transformed['calories'] / (df_transformed['carbs'] + df_transformed['fats'] + df_transformed['protein'] + 1e-6)

# Display the new features
df_transformed[['calories', 'protein', 'calorie_protein_ratio', 'calorie_density']].head()
```
